```python
data = pd.read_csv('../target-data/target-hospital-admissions.csv')
data['date'] = pd.to_datetime(data['date'])
data
```

```python
locations = data['location_name'].unique()
us_data = data[data['location_name'] == 'US']

fig, ax = plt.subplots()
ax.scatter(us_data['date'], us_data['value'], color='black', alpha=0.8, s=10)
ax.grid(color='gray', linestyle='--', linewidth=0.5, which='major')

# Set x-axis ticks to show only every nth label
xticks = us_data['date'][::8]  
ax.set_xticks(xticks)
ax.tick_params(axis='x', rotation=45)
ax.set_xlabel('Date')
ax.set_ylabel('Hospital Admissions')
ax.set_title('US')

plt.show()
```
